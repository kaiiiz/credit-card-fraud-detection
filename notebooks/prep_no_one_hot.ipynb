```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
train = pd.read_csv("../data/train.csv")
```

```python
test = pd.read_csv("../data/test.csv")
```

```python
train
```

```python
test
```

```python
les = {}
ss = StandardScaler()
mms = MinMaxScaler()

def preprocessing_train(train_df):
    labels = train_df["fraud_ind"]
    df = train_df.drop(["fraud_ind", "txkey"], 1) # unique value
    df = df.fillna("NA") # only flbmk & flg3dmsk has full value
    
    for c in df.keys():
        if not (df[c].dtype == np.int64 or df[c].dtype == np.float64):
            le = LabelEncoder()
            df.loc[:,c] = le.fit_transform(df.loc[:,c])
            les.update({c:le})

    x = ss.fit_transform(df)
    x = mms.fit_transform(x)
    df = pd.DataFrame(x, index=df.index, columns=df.columns)    
    
    df["fraud_ind"] = labels
    
    return df

train_df = preprocessing_train(train)
```

```python
train_df
```

```python
def preprocessing_test(test_df):
    labels = test_df["fraud_ind"]
    df = test_df.drop(["fraud_ind", "txkey"], 1)
    df = df.fillna("NA")
    
    for c in df.keys():
        if not (df[c].dtype == np.int64 or df[c].dtype == np.float64):
            df.loc[:,c] = les[c].transform(df.loc[:,c])
    
    x = ss.transform(df)
    x = mms.transform(x)
    df = pd.DataFrame(x, index=df.index, columns=df.columns)
    
    df["fraud_ind"] = labels
    
    return df

test_df = preprocessing_train(test)
```

```python
train_df.to_csv("../data/type1_train.csv", index=False)
```

```python
test_df.to_csv("../data/type1_test.csv", index=False)
```
